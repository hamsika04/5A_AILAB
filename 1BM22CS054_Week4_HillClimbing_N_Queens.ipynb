```python
import random

def count_attacks(queens):
    """Counts the number of pairs of queens that can attack each other."""
    total_attacks = 0
    size = len(queens)
    for row in range(size):
        for next_row in range(row + 1, size):
            if queens[row] == queens[next_row] or abs(queens[row] - queens[next_row]) == abs(row - next_row):
                total_attacks += 1
    return total_attacks

def generate_neighbors(queens):
    """Creates neighboring configurations by swapping positions of two queens."""
    neighboring_configs = []
    size = len(queens)
    for first in range(size):
        for second in range(first + 1, size):
            new_config = queens[:]
            new_config[first], new_config[second] = new_config[second], new_config[first]  # Swap positions
            neighboring_configs.append(new_config)
    return neighboring_configs

def climb_hill(initial_state):
    """Attempts to solve the N-Queens problem using the Hill Climbing method."""
    current_state = initial_state
    current_attacks = count_attacks(current_state)
    print(f"Starting configuration: {current_state} with {current_attacks} attacks")

    while True:
        neighbors = generate_neighbors(current_state)
        best_neighbor = None
        best_attacks = current_attacks

        for neighbor in neighbors:
            attacks = count_attacks(neighbor)
            if attacks < best_attacks:
                best_attacks = attacks
                best_neighbor = neighbor

        if best_attacks >= current_attacks:
            break  # No better configuration found, stop the search

        current_state = best_neighbor
        current_attacks = best_attacks
        print(f"Current best configuration: {current_state} with {current_attacks} attacks")

    return current_state, current_attacks

# Main execution block
try:
    size = int(input("Enter the number of queens (size of the board): "))
    if size <= 0:
        raise ValueError("Please enter a positive integer for the number of queens.")

    initial_input = input(f"Enter the row indices for the queens (space-separated, 0 to {size - 1}): ")
    initial_config = list(map(int, initial_input.split()))

    if len(initial_config) != size or any(pos < 0 or pos >= size for pos in initial_config):
        raise ValueError("Invalid input. Ensure you provide exactly {} indices in the range 0 to {}.".format(size, size - 1))

    final_solution, remaining_attacks = climb_hill(initial_config)
    if remaining_attacks == 0:
        print("Found a valid solution:")
        print(final_solution)
    else:
        print("No complete solution found. Best configuration with attacks:")
        print(final_solution, "with", remaining_attacks, "attacks.")
except ValueError as error:
    print(f"Input error: {error}")
```
