```python
def unify(term1, term2):
    # Check if the terms are the same (constants or identical terms)
    if term1 == term2:
        return {}

    # Handle case where term1 is a variable
    if is_variable(term1):
        return {term1: term2} if not occurs_in(term1, term2) else "FAILURE"

    # Handle case where term2 is a variable
    if is_variable(term2):
        return {term2: term1} if not occurs_in(term2, term1) else "FAILURE"

    # Ensure predicate symbols match
    if predicate(term1) != predicate(term2):
        return "FAILURE"

    # Ensure argument lengths match
    if len(arguments(term1)) != len(arguments(term2)):
        return "FAILURE"

    # Initialize substitution set
    subst = {}

    # Unify each argument recursively
    for arg1, arg2 in zip(arguments(term1), arguments(term2)):
        s = unify(apply_substitution(arg1, subst), apply_substitution(arg2, subst))
        if s == "FAILURE":
            return "FAILURE"
        subst.update(s)  # Merge substitutions

    return subst


# Helper functions
def is_variable(term):
    return isinstance(term, str) and term.islower()


def occurs_in(var, term):
    if var == term:
        return True
    elif isinstance(term, (list, tuple)):
        return any(occurs_in(var, t) for t in term)
    return False


def predicate(term):
    return term[0] if isinstance(term, list) else None


def arguments(term):
    return term[1:] if isinstance(term, list) else []


def apply_substitution(term, subst):
    if is_variable(term):
        return subst.get(term, term)
    elif isinstance(term, list):
        return [apply_substitution(arg, subst) for arg in term]
    return term


def expression_to_string(term):
    if isinstance(term, list):
        predicate_name = term[0]
        args = ", ".join(expression_to_string(arg) for arg in term[1:])
        return f"{predicate_name}({args})"
    return str(term)


# Test case
if __name__ == "__main__":
    term1 = ['Eats', 'x', 'Apple']
    term2 = ['Eats', 'Riya', 'y']

    result = unify(term1, term2)

    if result != "FAILURE":
        unified_term1 = apply_substitution(term1, result)
        print(expression_to_string(unified_term1))  # Expected: Eats(Riya, Apple)
    else:
        print("Unification Failed")
```
