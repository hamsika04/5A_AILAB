```python
import copy

# Function to find the position of the empty tile (0)
def find_blank_position(state):
    for i in range(len(state)):
        for j in range(len(state[i])):
            if state[i][j] == 0:
                return i, j

# Possible moves for the blank tile (up, down, left, right)
def get_possible_moves(state):
    moves = []
    row, col = find_blank_position(state)
    if row > 0:  # Move up
        moves.append((row - 1, col))
    if row < 2:  # Move down
        moves.append((row + 1, col))
    if col > 0:  # Move left
        moves.append((row, col - 1))
    if col < 2:  # Move right
        moves.append((row, col + 1))
    return moves

# Perform the move
def move_tile(state, new_row, new_col):
    row, col = find_blank_position(state)
    new_state = copy.deepcopy(state)
    new_state[row][col], new_state[new_row][new_col] = new_state[new_row][new_col], new_state[row][col]
    return new_state

# IDS helper to run DFS with a depth limit
def dfs_with_limit(state, goal, depth, visited):
    if state == goal:
        return True
    if depth == 0:
        return False

    visited.append(state)
    possible_moves = get_possible_moves(state)

    for move in possible_moves:
        new_state = move_tile(state, move[0], move[1])
        if new_state not in visited:
            if dfs_with_limit(new_state, goal, depth - 1, visited):
                return True
    visited.pop()
    return False

# Iterative Deepening Search
def ids(start, goal):
    depth = 0
    while True:
        visited = []
        if dfs_with_limit(start, goal, depth, visited):
            return visited
        depth += 1

# Function to display the puzzle state
def display_state(state):
    for row in state:
        print(row)
    print()

# Example usage
start_state = [
    [5, 4, 0],
    [6, 1, 8],
    [7, 3, 2]
]

goal_state = [
    [0, 1, 2],
    [3, 4, 5],
    [6, 7, 0]
]

if __name__ == '__main__':
    print("Initial State:")
    display_state(start_state)

    print("Goal State:")
    display_state(goal_state)

    result = ids(start_state, goal_state)
    if result:
        print("Solution Found! Steps:")
        for step in result:
            display_state(step)
    else:
        print("No solution exists!")
```

```python
from collections import deque

# Define the goal state
goal_state = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

# Define the possible moves (up, down, left, right) with corresponding move names
moves = {
    (0, 1): "Right",
    (0, -1): "Left",
    (1, 0): "Down",
    (-1, 0): "Up"
}

def is_goal(state):
    return state == goal_state

def dfs(state):
    visited = set()
    stack = deque([(state, [])])  # Initialize stack with initial state and empty path

    while stack:
        state, path = stack.pop()
        if is_goal(state):
            return path + [(state, "Goal")]  # Return the path to the goal state
        state_tuple = tuple(map(tuple, state))
        if state_tuple not in visited:
            visited.add(state_tuple)
            for i in range(3):
                for j in range(3):
                    if state[i][j] == 0:  # Find the position of the empty tile (0)
                        for move, direction in moves.items():
                            new_i, new_j = i + move[0], j + move[1]
                            if 0 <= new_i < 3 and 0 <= new_j < 3:
                                new_state = [row[:] for row in state]
                                new_state[i][j], new_state[new_i][new_j] = new_state[new_i][new_j], new_state[i][j]
                                # Add new state and updated path to stack, along with the direction
                                stack.append((new_state, path + [(state, direction)]))
    return None  # If no solution is found, return None

def print_solution(solution):
    if solution:
        print("Solution found:")
        for i, (state, direction) in enumerate(solution[1:]):  # Skip the initial state
            print(f"Move {i}: {direction}")
            for row in state:
                print(row)
            print()  # Print a newline for better readability
    else:
        print("No solution found")

# Example usage
initial_state = [[1, 2, 3], [4, 5, 6], [0, 7, 8]]
solution = dfs(initial_state)
print_solution(solution)
```
